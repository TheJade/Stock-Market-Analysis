```python
from scipy.fft import fft, ifft, fftfreq, rfft, rfftfreq
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
import yfinance as yf
from yahoofinancials import YahooFinancials

STOCK = "TSLA"
testing = False
INTERVAL = "1h"

tsla_df = yf.download(STOCK, period = "ytd", interval = INTERVAL)#start="2021-01-01", end="2021-04-15")
tsla_df.head()

length = len(tsla_df['Close'])
temp = tsla_df['Close']
temp2 = [0 for i in range(length)]

#testing
if testing:
    temp5 = [i*0.1 for i in range(length)]
    for i in range(length):
        temp[i] = np.sin(2 * np.pi * temp5[i]) + np.sin(3.3 * 2 * np.pi * temp5[i])

for i in range(length):
    temp2[i] = temp[i]

temp3 = rfft(temp2)

temp4 = rfftfreq(length, 1)

plt.figure(0)
plt.plot(temp2)
plt.figure(1)
plt.plot(temp4, np.abs(temp3))
plt.axis((0, 0.5, 0, 10000))
plt.title(STOCK + "   " + INTERVAL)
```
