```python
#testing the functionallity of matplotlib, math, and scipy.fft
#this is working as intended
from scipy.fft import fft, ifft, fftfreq, rfft, rfftfreq
import matplotlib.pyplot as plt
import numpy as np

length = 500
SAMPLE_RATE = 0.1

x = [i*SAMPLE_RATE for i in range(length)]
y = [0 for i in range(length)]

for i in range(length):
    y[i] = np.sin(2 * np.pi * x[i]) + np.sin(2 * 2 * np.pi * x[i])

plt.figure(0)
plt.plot(x, y)


plt.figure(1)
z = rfft(y)
x = rfftfreq(length, SAMPLE_RATE)
plt.plot(x, np.abs(z))

```
